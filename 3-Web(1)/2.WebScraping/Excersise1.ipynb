```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.imdb.com/chart/top/"
r = requests.get(url)
print("Status Code: ", r.status_code)
r.text[: 500]
```

```python
soup = BeautifulSoup(r.text, "html.parser")
```

```python
tag = "table"
attributes = {"class": "chart full-width"}
table_soup = soup.find(tag, attributes)
```

```python
table_data = []
for row in table_soup.find_all("tr"):
    row_text = []
    for ele in row.find_all("td"):
        row_text.append(ele.text.strip())
    table_data.append(row_text)

table_data
```

```python
import pandas as pd

list_df = pd.read_html(url)
list_df[0]
```

```python
df = pd.DataFrame(table_data)
```
