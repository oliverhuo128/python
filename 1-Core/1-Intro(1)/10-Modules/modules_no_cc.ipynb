## 13.1 Importing Functions from Modules

The Python language is typically distributed with the standard library, 
which provides additional types and functions that can be imported and used in our program. 

We use the `import` keyword to get access to the functions in a module.

We can then call these functions by giving the name of the module, followed by a dot, then the name of the function.

### Example: calling a function that returns the current working directory (folder)

The `os` module provides functions that interact with the operating system of the host computer. 

It has a function called `getcwd` that returns the current working directory (folder).

We can get this like so:

````python
import os

current_folder = os.getcwd()

print('current folder is ', current_folder)
````

## 13.2 Accessing Python Documentation 

The `help` function provides documentation on built-in functions and other Python objects.

For example, call `help(math.sin)` to show  help on this function. 

````python
help(math.sin)
````

We can read here that the input is assumed to be in radians, not degrees – we’ll use this fact when we write our driver code. Also, looking at this argument list, there’s an interesting final slash there – we will discover what these mean in a subsequent part of this module. 
Help is available on built-ins, too, e.g. `help(list)`. By the end of this module, you should start to feel confident with this level of technical detail, should you wish to.


## 13.3 Importing functions from packages

Packages are collections of modules. This allows libraries to scale up to provide big frameworks, yet still retain modularity between the different  modules. This is good because it breaks chains of dependencies, meaning a much leaner deployment workflow. Engineers can choose which modules from a package that they need.

For example, Matplotlib is an extensive data visualization package. There's one module within it, called pyplot, that provides some baseline plotting functions. 
if we just wanted to import that module, from the matplotlib package then we can write it like this:

```
import matplotlib.pyplot
```
Traditionally, for the `matplotlib.pyplot` module, we use an *alias* of `plt`, to make it quicker to type, each time we use it.

So we typically import it as
```
import matplotlib.pyplot as plt
```



````python
## Example: calling the plot function from matplotlib.pyplot 
import matplotlib.pyplot as plt

y_values = [1,3,2,4,3,5]
plt.plot(y_values)
````

### Concept Check: Plotting Sine and Cosine Curves (*)

- Make an empty list of `y_values`
- Iterate over a sequence of numbers from 0 to 360 (these will be our x-values)
- Each iteration, convert the number from degrees to radians (i.e. divide by 2 pi)
- Take the sine of the angle, and append this to the list of `y_values` 
- After the iteration, plot the `y_values`.
- Use the `help` function to access the documentation for `plt.plot`. Use the information on the first page to add 'red plusses' to your plot.
- Include the line `plt.savefig('trig.png')` to save this plot as an image file called `'trig.png'`
- Write additional code to include a cosine plot in your figure.
- Copy your code into the `modules_ex1` file inside the `Exercises` folder, so that when you run `python modules_ex1.py`, it creates the image file `trig.png`. 


````python
import math
y_values = []

for x_deg in range(361):
    x_rad = x_deg / (2 * math.pi)
    y_values.append(math.sin(x_rad))

plt.plot(y_values, "r+")

plt.savefig("trig_cos.png")
````
