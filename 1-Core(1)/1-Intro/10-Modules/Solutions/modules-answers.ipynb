### Concept Check: Plotting Sine and Cosine Curves

- Make an empty list of `y_values`
- Iterate over a sequence of numbers from 0 to 360 (these will be our x-values)
- Each iteration, convert the number from degrees to radians (i.e. divide by 2 pi)
- Take the sine of the angle, and append this to the list of `y_values` 
- After the iteration, plot the `y_values`.
- Repeat this to make a cosine plot 
- Use the `help` function to access the documentation for `plt.plot`. Use the information on the first page to add 'red plusses' to the sine plot.

```
import matplotlib.pyplot as plt
y_values = []
for degree in range(360):
    radian = degree*2*math.pi/360
    y_values.append(math.sin(radian))
plt.plot(y_values, 'r+')

y_values = []
for degree in range(360):
    radian = degree*2*math.pi/360
    y_values.append(math.cos(radian))
plt.plot(y_values)

```

````python
import matplotlib.pyplot as plt
import math
y_values = []
for degree in range(360):
    radian = degree*2*math.pi/360
    y_values.append(math.sin(radian))
plt.plot(y_values, 'r+')

y_values = []
for degree in range(360):
    radian = degree*2*math.pi/360
    y_values.append(math.cos(radian))
plt.plot(y_values)
````
